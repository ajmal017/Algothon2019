```python
df = pd.read_csv('price_data/TTC.csv')
#(df.iloc[0:,[0,4]]).plot()
#(df.iloc[0:,[0,11]]).plot()
```

```python
X=df.iloc[0:,11]
size = int(len(X) * 0.66)
X=list(X)
train, test = X[0:size], X[size:len(X)]
predictions = list()
obs=list()
```

```python
for t in range(len(test)):
    model = ARIMA(train, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs=test[t]
    train.append(obs)
```
